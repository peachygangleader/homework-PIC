```python
import pandas as pd
import matplotlib.pyplot as plt

txt = pd.read_csv("file.txt")
txt
```

```python
plt.plot(txt.n, txt.t_pr, label='Метод квадратур')
plt.plot(txt.n, txt.t_tr, label='Метод трапеций')
plt.plot(txt.n, txt.t_mc, label='Метод Монте-Карло')
plt.legend()
plt.ylabel('Время, мкс')
plt.xlabel('Число разбиений')
plt.show()
```

```python
plt.plot(txt.n, txt.eps_pr, label='Метод квадратур')
plt.plot(txt.n, txt.eps_tr, label='Метод трапеций')
plt.plot(txt.n, txt.eps_mc, label='Метод Монте-Карло')
plt.legend()
plt.ylabel('Относительная ошибка')
plt.xlabel('Число разбиений')
plt.semilogy()
plt.show()
```
